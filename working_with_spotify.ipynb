```python
def search_song(x):
    if isinstance(x, str):
        results_search_song = sp.search(q="track:"+x, limit=50)
    elif isinstance(x, list):
        y = ""
        for i, search in enumerate(x):
            if i < len(x)-1:
                y += search+", "
            else:
                y += search
        results_search_song = sp.search(q="track:"+y, limit=50)
    song_id = []
    counter = 1
    for i in results_search_song["tracks"]["items"]:
        print("-------------------------------")
        print("Number:\t", str(counter))
        print("artist:\t",i["album"]["artists"][0]["name"])
        print("song:\t",i["name"])
        print("album:\t",i["album"]["name"])
        counter += 1
        song_id.append(i["id"])
    if len(song_id) == 0:
        print("No matches")
        return
    index = input("Please enter Number: ")
    clear_output()
    result_track = sp.track(song_id[int(index)-1])
    print("-------------------------------")
    print("artist:\t",result_track["album"]["artists"][0]["name"])
    print("song:\t",result_track["name"])
    print("album:\t",result_track["album"]["name"])
    print("id:\t",result_track["id"])
    print("href:\t",result_track["href"])
    print("uri:\t",result_track["uri"])
    
    return song_id

    
```

```python
def get_audio_features(list_of_songs):
    audio_features = {
    'songname': [],
    'artist': [],
    'album': [],
    'danceability': [],
    'energy': [],
    'key': [],
    'loudness': [],
    'mode': [],
    'speechiness': [],
    'acousticness': [],
    'instrumentalness': [],
    'liveness': [],
    'valence': [],
    'tempo': [],
    'id': [],
    'uri': [],
    'track_href': [],
    'duration_ms': []    
    }
    for item in list_of_songs:
        feature = sp.audio_features([item])
        result_track = sp.track(item)
        audio_features['songname'].append(result_track["name"])
        audio_features['artist'].append(result_track["album"]["artists"][0]["name"])
        audio_features['album'].append(result_track["album"]["name"])
        audio_features['danceability'].append(feature[0]['danceability'])
        audio_features['energy'].append(feature[0]['energy'])
        audio_features['key'].append(feature[0]['key'])
        audio_features['loudness'].append(feature[0]['loudness'])
        audio_features['mode'].append(feature[0]['mode'])
        audio_features['speechiness'].append(feature[0]['speechiness'])
        audio_features['acousticness'].append(feature[0]['acousticness'])
        audio_features['instrumentalness'].append(feature[0]['instrumentalness'])
        audio_features['liveness'].append(feature[0]['liveness'])
        audio_features['valence'].append(feature[0]['valence'])
        audio_features['tempo'].append(feature[0]['tempo'])
        audio_features['id'].append(feature[0]['id'])
        audio_features['uri'].append(feature[0]['uri'])
        audio_features['track_href'].append(feature[0]['track_href'])
        audio_features['duration_ms'].append(feature[0]['duration_ms'])
        
    audio_features_df = pd.DataFrame(audio_features)
        
    return audio_features_df
```
